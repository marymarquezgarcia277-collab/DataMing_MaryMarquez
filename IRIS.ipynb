```python
import pandas as pd
df = pd.read_csv("iris.csv")
df.head()
```

# **Primer análisis**

```python
df["species"].value_counts() #Cantidad de flores por especie
```

# **Segundo análisis (promedios)**

```python
df.groupby("species").mean()
```

# **Gráfico de dispersión (Scatter Plot)**
Interpretación

Este gráfico muestra la relación entre el largo y el ancho del pétalo.

Explicación

Se observan grupos de puntos separados, lo que indica que las diferentes especies de flores tienen tamaños de pétalos distintos.

Insight accionable

El tamaño del pétalo puede utilizarse como una característica clave para clasificar las especies de iris.

```python
import matplotlib.pyplot as plt
plt.scatter(df["petal_length"], df["petal_width"])
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Relación entre largo y ancho del pétalo")
plt.show()
```

# **Gráfico de barras**
Interpretación

El gráfico muestra cuántas flores hay de cada especie.

Explicación

Las tres especies tienen la misma cantidad de observaciones (50 cada una).

Insight

El dataset está balanceado, lo cual es ideal para realizar análisis y modelos de clasificación.

```python
df["species"].value_counts().plot(kind="bar")
plt.title("Cantidad de flores por especie")
plt.xlabel("Especie")
plt.ylabel("Cantidad")
plt.show()
```

# **Histograma**

Interpretación

El histograma muestra cómo se distribuyen los valores del largo del pétalo.

Explicación

Se observan diferentes rangos de valores que pueden corresponder a distintas especies.

Insight

El largo del pétalo parece ser una variable importante para diferenciar especies.



```python
df["petal_length"].plot(kind="hist", bins=20)
plt.title("Distribución del largo del pétalo")
plt.xlabel("Petal Length")
plt.show()
```

# **Boxplot**
Interpretación

El boxplot compara la distribución del largo del pétalo entre especies.

Explicación

Algunas especies tienen pétalos claramente más largos que otras.

Insight

Esto confirma que el tamaño del pétalo es un buen indicador para distinguir especies.

```python
import seaborn as sns

sns.boxplot(x="species", y="petal_length", data=df)
plt.title("Distribución del largo del pétalo por especie")
plt.show()
```
